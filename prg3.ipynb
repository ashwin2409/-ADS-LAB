```python
l1=['good','fine','nice','happy','positive']
l2=['sad','bad','tried','frustrated','negative','not']
str1=input("Enter your Response")
flag=0
ncount=0
t=str1.split()
for i in range(len(t)):
  for j in range(len(l1)):
    if t[i]==l1[j]:
      flag=1
  for k in range(len(l2)):
    if t[i]==l2[k]:
     ncount=ncount+1
     flag=1
if flag==0:
 print("your are in another mood")
elif ncount%2==0:
  print("positive")
else:
  print("negative")
```
